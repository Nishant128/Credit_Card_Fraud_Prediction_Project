## Splitting the data 

```python
X=credit_data.iloc[:,:-1]
Y=credit_data.iloc[:,-1]
```

```python
X
```

```python
Y
```

## Training and Testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=2,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model

```python
logestic_model=LogisticRegression()
```

```python
logestic_model.fit(X_train,Y_train)
```

### X_train Prediction

```python
X_train_prediction=logestic_model.predict(X_train)
X_train_accuracy=accuracy_score(X_train_prediction,Y_train)
X_train_accuracy
```

### X_test Prediction

```python
X_test_prediction=logestic_model.predict(X_test)
X_test_accuracy=accuracy_score(X_test_prediction,Y_test)
X_test_accuracy
```
